```python
data = sns.load_dataset('tips')  # tips dataset already in seaborn library
```

```python
data.head()
```

```python
plt.plot(data['total_bill'])  # plotting data on graph
plt.show()
```

```python
data.describe()
```

```python
sns.lineplot(data['total_bill'])
plt.show()
```

```python
sns.scatterplot(data['total_bill'])
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data)   # 
```

```python
data.describe()
```

```python
sns.histplot(data['total_bill'])  # plotting histogram of dataset feature
```

```python
plt.hist(data['total_bill'])
plt.show()
```

```python
sns.displot(data['total_bill'])
```

```python
sns.kdeplot(data['total_bill'])

# A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram
```

```python
sns.relplot(x='total_bill',y='tip',data=data) # relation plot b/w x and y 
```

```python
data['time'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',data=data)

# In Seaborn, the hue parameter is used to color-code data points or lines based on a categorical variable, 
# allowing for the visualization of multiple groups or categories within a single plot. 
```

```python
data.head()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',data=data)

# The Seaborn Relational Plot (relplot) allows us to visualise how variables within a dataset relate to each other
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',col='smoker',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',col='smoker',kind='line',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',row='smoker',kind='line',data=data)
```

```python
data.head()
```

```python
data['day'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',col='day',kind='scatter',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',col='day',kind='line',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',data=data)
plt.xlabel('X-axis total_bill')
plt.ylabel('Y-axis tip')
plt.title("Total_bill vs Tip Scatter Plot")

```
