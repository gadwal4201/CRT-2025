```python
x_point = np.array([1,5,8,7])
plt.plot(x_point,'o:b',ms=30)  # ms=30 increase the marker size
plt.show()
```

```python
x_point = np.array([1,5,8,7])
plt.plot(x_point,marker='*',ms=20,mec='r')  # color change
plt.show()
```

```python
x_point = np.array([1,5,8,7])
plt.plot(x_point,marker='*',ms=30,mec='g',mfc='r')  # mec color the boundary,  mfc color inside of symbols
plt.show()
```

```python
x_point = np.array([1,5,8,7])
plt.plot(x_point,ls='dotted')  # ls --> line style is dotted
plt.show()
```

```python
arr1 = np.array([88,354,4,4,35,4,64,4,4])
arr2 = np.array([99,546,45,3541,74,341,87,94,7])
plt.scatter(arr1,arr2)
plt.show()
```

```python
arr1 = np.array([88,34,95,4,35,4,64,100,11])
arr2 = np.array([99,56,45,35,74,31,87,94,7])
plt.scatter(arr1,arr2)
arr3 = np.array([88,54,54,6,9,6,4])
arr4 = np.array([64,4,51,4,86,94,7])
plt.scatter(arr3,arr4)
plt.show()
```

```python
arr1 = np.array([88,34,95,4])
arr2 = np.array([99,56,45,35])
c= np.array(['red','green','blue','yellow'])
plt.scatter(arr1,arr2,c = c)

plt.show()        
```

```python
arr1 = np.array([88,34,95,4])
arr2 = np.array([99,56,45,35])
colour= np.array(['red','green','blue','yellow'])

size= np.array([100,200,300,400])
plt.scatter(arr1,arr2,s=size,c=colour)

plt.show()        
```

```python
arr1 = np.array([88,34,95,4])
arr2 = np.array([99,56,45,35])
colour= np.array(['red','green','blue','yellow'])

size= np.array([100,200,300,400])
plt.scatter(arr1,arr2,s=size,c=colour,alpha=0.4)

plt.show()        
```

```python
arr1 = np.array([88,34,95,4])
arr2 = np.array([99,56,45,35])
colour= np.array(['red','green','blue','yellow'])

size= np.array([100,200,300,400])
plt.scatter(arr1,arr2,s=size,c=colour,alpha=0.5,cmap='nipy_spectral')

plt.show()        
```

```python
arr1 = np.array([88,34,95,4])
arr2 = np.array([99,56,45,35])
colour= np.array(['red','green','blue','yellow'])

size= np.array([100,200,300,400])
plt.scatter(arr1,arr2,s=size,c=colour,alpha=0.5,cmap='nipy_spectral')
plt.colorbar()

plt.show()        
```
