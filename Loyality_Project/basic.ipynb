```python
dt = { 
    'Loyal Customer' : 1,
    'disloyal Customer' : 0
    
    }

df["Customer Type"]=df["Customer Type"].map(dt)
```

```python
df["Customer Type"].unique()
```

```python
df["Type of Travel"].unique()
```

```python
dt1 = {
    'Business travel' : 1,
    'Personal Travel' : 0
}
df["Type of Travel"]=df["Type of Travel"].map(dt1)
df["Type of Travel"].unique()
```

```python
dt2 = {
    'neutral or dissatisfied' : 1,
    'satisfied' : 0
}
df["satisfaction"]=df['satisfaction'].map(dt2)
```

```python
df['Class'].unique()
```

```python
dt3 = { 
    'Eco Plus' : 0,
    'Business' : 1,
    'Eco' : 2
}
df['Class']=df['Class'].map(dt3)
```

```python
df['Gender'].unique()
```

```python
dt4 = {
    'Male' : 1,
    'Female' : 0
}
df['Gender']=df['Gender'].map(dt4)
```

```python
df.drop('satisfaction',axis=1)
```

```python
x = df.drop('satisfaction',axis=1)

y = df['satisfaction']
```
